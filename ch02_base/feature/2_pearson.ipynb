```python
import pandas as pd
```

```python
advertising = pd.read_csv("../../data/advertising.csv")
print(advertising.head())
print(advertising.describe())
print(advertising.shape)
```

```python
# 数据预处理
# 去掉ID
advertising = advertising.drop(advertising.columns[0],axis=1)
print(advertising.head())
# 去掉空值
advertising = advertising.dropna()
print(advertising.head())
# 提取特征和标签
X = advertising.drop("Sales",axis=1)
Y = advertising["Sales"]

print(X.shape)
print(Y.shape)




```

```python
# 计算皮尔逊相关系数
print(X.corrwith(Y,method="pearson"))
```

```python
print(advertising.corr(method="pearson"))
```

```python
# 将相关系数矩阵画成热力图
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(advertising.corr(),cmap="coolwarm",fmt=".2f",annot=True)
plt.title("Feature Correlation Matrix")
```
